```julia
draw = [1] #precisa criar um vetor unitário para que o índice seja mutável

function rand_fixo!(U, draw)
    #tira um draw do U pré-determinado
    #atualiza o índice i para que o próximo draw seja num índice diferente
    
    
    u = U[draw]
    
    draw[1] += 1
    
    return u[1]
end
    
```

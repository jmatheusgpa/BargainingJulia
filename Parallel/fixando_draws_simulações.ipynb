```julia
draw = [1] #precisa criar um vetor unitário para que o índice seja mutável

function rand_fixo!(U, draw)
    #tira um draw do U pré-determinado
    #atualiza o índice i para que o próximo draw seja num índice diferente
    
    
    u = U[draw]
    
    draw[1] += 1
    
    return u[1]
end
    
```

```julia
S = 100
N = 66
ndraws = 120

U = rand(S, N, ndraws)
```

```julia
function draw_rand!(U, draw)
    
    u = U[draw]
    
    draw[1] += 1
    
    return u[1]
    
end
```

```julia
function simulate(U)
    
    
    draw = [1]
    
    for i in 1:5
        @show rand_fixo!(U, draw)
        @show draw
        
        @show draw_rand!(U, draw)
        @show draw
        
        @show draw_rand!(U, draw)
        @show draw
    end
    
    
end
```
